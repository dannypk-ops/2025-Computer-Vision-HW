## Image Addition

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 1.   ( )
add_img_01 = cv2.imread("images/add_01.png", cv2.IMREAD_COLOR)
add_img_02 = cv2.imread("images/add_02.png", cv2.IMREAD_COLOR)

# 2.   
add_img = cv2.add(add_img_01, add_img_02)

# matplotlib   
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.title("Image 1")
plt.imshow(cv2.cvtColor(add_img_01, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Image 2")
plt.imshow(cv2.cvtColor(add_img_02, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Added Image")
plt.imshow(cv2.cvtColor(add_img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Image Subtraction

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 1.   ( )
sub_img_01 = cv2.imread("images/sub_01.png", cv2.IMREAD_COLOR)
sub_img_02 = cv2.imread("images/sub_02.png", cv2.IMREAD_COLOR)

# 2.   
sub_img = cv2.subtract(sub_img_01, sub_img_02)

# matplotlib   
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.title("Image 1")
plt.imshow(cv2.cvtColor(sub_img_01, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Image 2")
plt.imshow(cv2.cvtColor(sub_img_02, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Subtracted Image")
plt.imshow(cv2.cvtColor(sub_img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Image Multiplication

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 1.   ( )
mul_img_01 = cv2.imread("images/mul_01.png", cv2.IMREAD_COLOR).astype(np.float32)

# 2.    (  1.5)
bright_mask = np.full(mul_img_01.shape, 1.3, dtype=np.float32)

# 3.   
mul_img = cv2.multiply(mul_img_01, bright_mask)

# 4.  0~255      uint8 
mul_img = np.clip(mul_img, 0, 255).astype(np.uint8)

# matplotlib   
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(mul_img_01.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Brightened Image")
plt.imshow(cv2.cvtColor(mul_img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Image Division
