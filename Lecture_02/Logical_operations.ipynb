## Logical Operations

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 1.  
img_01 = cv2.imread('images/div_01.jpeg')
img_02 = cv2.imread('images/div_02.jpeg')


# 2. bilinearization
_, bin_01 = cv2.threshold(img_01, 127, 255, cv2.THRESH_BINARY)
_, bin_02 = cv2.threshold(img_02, 127, 255, cv2.THRESH_BINARY)

# 2.  
and_img = cv2.bitwise_and(bin_01, bin_02)
or_img = cv2.bitwise_or(bin_01, bin_02)
xor_img = cv2.bitwise_xor(bin_01, bin_02)
not_img_01 = cv2.bitwise_not(bin_01)
not_img_02 = cv2.bitwise_not(bin_02)

# 3.   
titles = ['Image 01', 'Image 02', 'AND', 'OR', 'XOR', 'NOT Image 01', 'NOT Image 02']
images = [bin_01, bin_02, and_img, or_img, xor_img, not_img_01, not_img_02]

# 4. 
plt.figure(figsize=(14, 8))
cols = 4
rows = int(np.ceil(len(images) / cols))

for i, (title, img) in enumerate(zip(titles, images)):
    plt.subplot(rows, cols, i + 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
